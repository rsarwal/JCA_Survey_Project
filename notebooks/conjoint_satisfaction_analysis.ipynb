```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Load the dataset
file_path = '../data/survey_analysis_full.csv'
df = pd.read_csv(file_path)
df.head()
```

```
# Average ratings for membership features
features = [
    'pref_free_guest_pass', 'pref_event_access', 
    'pref_priority_seating', 'pref_discounted_merch'
]
avg_features = df[features].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
avg_features.plot(kind='bar', color='skyblue')
plt.title('Average Preference Ratings for Membership Features')
plt.ylabel('Average Rating (1-5)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```
# Satisfaction by membership status
satisfaction = [
    'satisfaction_events', 'satisfaction_communication', 'satisfaction_support'
]
df_grouped = df.groupby('membership_status')[satisfaction].mean()

plt.figure(figsize=(8, 6))
df_grouped.plot(kind='bar')
plt.title('Average Satisfaction by Membership Status')
plt.ylabel('Satisfaction (1-5)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```
# Satisfaction by age group
age_grouped = df.groupby('age_group')[satisfaction].mean()

plt.figure(figsize=(8, 6))
age_grouped.plot(kind='line', marker='o')
plt.title('Satisfaction by Age Group')
plt.ylabel('Satisfaction (1-5)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
